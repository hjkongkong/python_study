## 컬럼의 이름 다루기
데이터 프레임의 각종 정보들을 변경할 때

```python
# 컬럼의 이름은 renew 함수를 활용해서 바꿀 수 있다
CCTV_Seoul.rename(columns={'기관명':'구별'},inplace=True)
```

## inplace 매개변수의 의미
*  inplace가 True면 데이터 프레임 자체를 변경
* inplace가 False면 변경된 데이터 프레임을 새롭게 생성 **(false가 기본값)**


# 엑셀 파일 불러오기

```python
pop_Seoul = pd.read_excel('./data/01. population_in_Seoul.xls', encoding='utf-8')
pop_Seoul.head()
```

원하지 않는 곳에 있는 데이터도 불러와 진다.
* 행 데이터를 따로 걸러 낼 것이다 -> header
* 열 데이터를 따로 걸러 낼 것이다 -> useclos

```python
pop_Seoul = pd.read_excel('./data/01. population_in_Seoul.xls',
                          header = 2, # 엑셀상의 2번째 행 부터 데이터를 읽어 올 것-> 세 번째 행부터 데이터를 읽어온다.
                          usecols='B, D, G, J, N',
                          encoding='utf-8')
pop_Seoul.head()
```

## 인구 데이터 컬럼명 바꿔주기

```python
pop_seoul_columns = pop_Seoul.columns
pop_Seoul.rename(columns={
    pop_seoul_columns[0] : '구별',
    pop_seoul_columns[1] : '인구수',
    pop_seoul_columns[2] : '한국인',
    pop_seoul_columns[3] : '외국인',
    pop_seoul_columns[4] : '고령자'
},inplace=True)

pop_Seoul.head()
```

# 구별 CCTV, 인구수 데이터를 이용해 데이터 분석하기
1. 가설
    - 인구수가 많으면 CCTV가 많을까?
2. 비즈니스에 적용
    - 인구수에 비해 CCTV가 없으면 CCTV를 더 설치한다는 비즈니스를 세울 수 있다.
    - 인구수에 비해 CCTV가 많으면 왜 많은 지를 생각해 볼 수 있다.

# CCTV 데이터 프레임 정렬하기

```python
CCTV_Seoul.head()
```

```python
# 소계를 기준으로 내림차순 정렬
CCTV_Seoul.sort_values(by='소계', ascending=False).head()
```

# 데이터 분석을 할 때 의미있는 데이터는 비율 데이터이다.
소계, n 년도 CCTV 대수같은 절대적인 수치인 데이터보다, 대비를 한 빈율 데이터가 훨씬  유리하다.

```python
# CCTV 증가율을 구하기
# {(2014년 + 2015년 + 2016년)/2013년도 이전}+100

CCTV_Seoul['최근증가율'] = (CCTV_Seoul['2014년']+CCTV_Seoul['2015년']+CCTV_Seoul['2016년'])/CCTV_Seoul['2013년도 이전']*100
```

```python
CCTV_Seoul.head()
```

### DataFrame에서 행을 삭제하는 방법

```python
pop_Seoul['구별'].unique() # 구별 컬럼의 데이터를 중복없이 보여준다. 마치 set처럼
```

```python
pop_Seoul['외국인비율'] = (pop_Seoul['외국인']/pop_Seoul['인구수'])*100
pop_Seoul['고령자비율'] = (pop_Seoul['고령자']/pop_Seoul['인구수'])*100
pop_Seoul.head()
```

## CCTV와 인구수 데이터에 대한 분석 정리
1. 인구수가 제일 많은 top 5 지역은 송파구, 강서구, 강남구, 노원구, 관악구
2. 외국인이 제일 많은 top 5 지역은 영등포구, 구로구, 금천구, 관악구, 동대문구
3. 외국인 비율이 제일 높은 top 5 지역은 강서구, 송파구, 은평구, 노원구, 관악구
4. 고령자가 제일 많은 top 5 지역은 강서구, 송파구, 은평구, 노원구, 관악구
5. 고령자 비율이 제일 높은 top 5 지역은 강북구, 종로구, 중구, 용산구, 서대문구

========
1. 영등포구가 제일 외국인이 많은데, 인접한 구로구와 금천구에도 외국인이 굉장히 많이 거주한다.
2. 송파구는 고령자의 수가 제일 많지만 인구수 대비 비율(고령자비율)은 꽤나 낮다. 하지만 강북구, 종로구, 중구, 용산구, 서대문구는 고령자 비율이 높기 때문에 노령화가 우려됩니다.
