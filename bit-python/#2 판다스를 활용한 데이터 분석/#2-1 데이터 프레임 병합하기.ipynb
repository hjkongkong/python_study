```python
import pandas as pd
```

```python
# 딕셔너리를 이용해서 데이터 프레임 만들기
df1 = pd.DataFrame({'A':['A0','A1','A2','A3'],
                   'B':['B0','B1','B2','B3'],
                   'C':['C0','C1','C2','C3'],
                   'D':['D0','D1','D2','D3']})

df2 = pd.DataFrame({'A':['A4','A5','A6','A7'],
                   'B':['B4','B5','B6','B7'],
                   'C':['C4','C5','C6','C7'],
                   'D':['D4','D5','D6','D7']})

df3 = pd.DataFrame({'A':['A8','A9','A10','A11'],
                   'B':['B8','B9','B10','B11'],
                   'C':['C8','C9','C10','C11'],
                   'D':['D8','D9','D10','D11']})
```

```python
df1
```

```python
df2
```

```python
df3
```

# 열 방향으로 데이터 합치기
데이터 프레임을 위에서 아래로 합치기(세로방향 합치기)

```python
result = pd.concat([df1,df2,df3])
result
```

keys 옵션을 이용해 데이터 프레임마다 고유 인덱스를 따로 부여할 수 있다

```python
result = pd.concat([df1,df2,df3],keys=['x','y','z'])
result
```

```python
result.index # 튜플형태로 인덱스 등장
```

```python
# 레벨에 따라서 인덱스 확인하기
result.index.get_level_values(0)
```

```python
result.index.get_level_values(1)
```

서로 다른 인덱스와 컬럼을 가진 데이터 프레임 합치기

```python
df4 = pd.DataFrame({'B':['B2','B3','B6','B7'],
                   'D':['D2','D3','D6','D7'],
                   'F':['F2','F3','F6','F7']}, index=[2,3,6,7])
df4
```

df1과 df4를 합치기<br>
axis 옵션을 이용하면 행과 열 방향을 설정할 수 있다.

* axis가 0이면 세로병합 - 행을 늘린다(아래쪽으로)
* axis가 1이면 가로병합 - 열을 늘린다.(컬럼을 늘린다)

```python
result = pd.concat([df1,df4],axis=1)
result
```

합쳐지는 데이터 프레임끼리 공통된 인덱스가 있다면 해당 인덱스의 열에 자연스럽게 들어간다. 인덱스가 없다면 NaN으로 처리된다.

```python
result = pd.concat([df1,df4]) # axis의 기본값은 0
result
```

합치는 기준을 정의하는 join 매개변수

```python
# RDBMS의 inner JOIN
# pandas의 concat에서는 공통된 인덱스만 출력
result = pd.concat([df1,df4],join='inner',axis=1)
result
```

reindex 옵션을 활용하면 기준 데이터 프레임의 인덱스 저장 가능
* 합쳐지는 데이터 프레임에 기준 데이터 프레임과 동일한 인덱스가 없으면 해당 행(row)는 버린다.

```python
result = pd.concat([df1,df4], axis=1).reindex(df1.index) # reindex 함수에 기준이 될 인덱스의 목록을 넣어준다
result
```

데이터 프레임이 합쳐지면서 새로운 인덱스를 부여할 때 ignore_index

```python
result = pd.concat([df1,df4], ignore_index=True)
result
```

concat으로 데이터 프레임을 합치는 방법은 index와 연관이 깊다.<br>
# 컬럼을 기준으로 합치게 해주는 merge
concat은 공통된 키가 되는 칼럼이 없을 때 사용이 가능하다. merge는 공통 칼럼이 없으면 사용이 불가능하다.

```python
left = pd.DataFrame({'key':['K0','K4','K2','K3'],
                   'A':['A0','A4','A2','A3'],
                   'B':['B0','B4','B2','B3']})
right = pd.DataFrame({'key':['K0','K4','K2','K3'],
                   'C':['C0','C4','C2','C3'],
                   'D':['D0','D4','D2','D3']})
```

```python
left
```

```python
right
```

```python
pd.merge(left,right,on="key") # 서로 동시에 존재하는 컬럼인 key를 기준으로 가로방향으로 합쳐진다.
```

별다른 설정을 하지 않으면 공통되지 않은 key에 대한 row는 삭제가 된다.<br><br>
how 옵션을 활용하여 기준이 되는 데이터 프레임을 지정할 수도 있다.

```python
pd.merge(left,right,how='left',on='key') #how의 left는 말그대로 left임.데이터가 아니라
```

```python
pd.merge(left,right,how='right',on="key")
```

```python
pd.merge(left,right,how='outer',on="key")
```
