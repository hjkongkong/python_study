### seaborn의 set_style에 대하여
* set_style은 seaborn 및 matplotlib을 초기화
* set_style을 호출하면 설정들이 모두 초기화 되기 때문에, 한글개짐 방지를 우ㅏ한 한글 폰트를 불러오는 코드도 다시 처음부터 불러와야 한다.
* 위에서 한글이 잘 표시되다가 갑자기 안나오면 set_style 문제가 있는 경우가 많다.

```python
# 처음이라고 생각하고 처음부터 불러오기
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style('whitegrid')
%matplotlib inline
```

```python
# seaborn에 있는 샘플 데이터셋 로드
tips = sns.load_dataset('tips') # 식당 데이터셋
tips.head()
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(x="day", y="total_bill",data=tips)
# 최저, 최고값, 평균, 분산, 표준편차 범위 => 통계적인 의미를 시각화 할 때 사용
plt.show()
```

hue 옵션을 사용하면 데이터를 분류해서 시각화 할 수 있다.

ex) 요일별 지불금액을 확인하는데, 흡연자는 얼마를 냈을까? 비흡연자는 얼마나 냈지?

```python
plt.figure(figsize=(10,6))
sns.boxplot(x="day", y="total_bill",data=tips, hue="smoker", palette='Set3',)
plt.show()
```

# 산포도( scatter) 및 회귀까지 볼 수 있는 lmplot

```python
sns.set_style('darkgrid')
sns.lmplot(x='total_bill', y='tip',data=tips,size=7) # 지불금액이 x일 때 팁(y)은 보통 얼마를 냈는가?
plt.show()
```

```python
sns.lmplot(x='total_bill', y='tip',data=tips,hue='smoker',size=7) # 지불금액이 x일 때 팁(y)은 보통 얼마를 냈는가?
plt.show()
```

# 각 데이터의 크기를 색상의 농도로 표시할 때 사용이 가능
* Heat맵을 이용할 때는 숫자로 이루어진 데이터셋에서 활용해야 한다. (~별 컬럼들의 농도)

```python
flights = sns.load_dataset('flights')
flights.head()
```

```python
# month을 기준으로 월별 승객 인원수가 기준이 되는 그래프
# month를 인덱스로 삼고, 항목(칼럼)을 연도(year) -> 월별 탑승객 수를 확인. 연도는 항목
flights = flights.pivot("month", 'year','passengers') # 인덱스, 컬럼, 데이터
flights.head()
```

```python
flights
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(flights,annot=True,fmt='d') # annot = 히트맵에 숫자 표시 여부, fmt : 정수로 표현할 지, 실수로 표현할지 결정
plt.show()
```
