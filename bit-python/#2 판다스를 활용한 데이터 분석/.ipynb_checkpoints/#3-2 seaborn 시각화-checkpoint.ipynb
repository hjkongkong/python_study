# Seaborn은 Matplotlib을 예쁘게 꾸며준다는 느낌으로 사용한다.
* 

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.linspace(0,14,100)

y1 = np.sin(x)
y2 = 2*np.sin(x*0.5)
y3 = 3*np.sin(x*1.0)
y4 = 4*np.sin(x*1.5)
```

```python
plt.figure(figsize=(10,6))
plt.plot(x,y1 ,x,y2, x,y3, x,y4)
plt.show()
```

```python
# seaborn 활용을 위해서는 반드시 matplotlib을 import한다.
import seaborn as sns
```

seaborn으로 스타일 등을 설정하면 matplotlib도 그 영향을 받는다

```python
sns.set_style('whitegrid')# seaborn의 whitegrid 스타일 적용

plt.figure(figsize=(10,6)) # figure를 호출하게 되면 새로운 시각화 그래프를 그리기 위한 환경을 새로 만들게 된다.
plt.plot(x,y1 ,x,y2, x,y3, x,y4)
plt.show()
```

### seaborn의 set_style에 대하여
* set_style은 seaborn 및 matplotlib을 초기화
* set_style을 호출하면 설정들이 모두 초기화 되기 때문에, 한글개짐 방지를 우ㅏ한 한글 폰트를 불러오는 코드도 다시 처음부터 불러와야 한다.
* 위에서 한글이 잘 표시되다가 갑자기 안나오면 set_style 문제가 있는 경우가 많다.

```python
# 처음이라고 생각하고 처음부터 불러오기
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style('whitegrid')
%matplotlib inline
```

```python
# seaborn에 있는 샘플 데이터셋 로드
tips = sns.load_dataset('tips') # 식당 데이터셋
tips.head()
```
