```python
import pandas as pd # pandas 패키지
```

```python
pop_Seoul = pd.read_excel('./data/01. population_in_Seoul.xls',
                          header = 2, # 엑셀상의 2번째 행 부터 데이터를 읽어 올 것-> 세 번째 행부터 데이터를 읽어온다.
                          usecols='B, D, G,H,I, J, K,L',
                          encoding='utf-8')
pop_Seoul.head()
```

```python
pop_seoul_columns = pop_Seoul.columns
pop_Seoul.rename(columns={
    pop_seoul_columns[0] : '구별',
    pop_seoul_columns[1] : '총인구',
    pop_seoul_columns[2] : '한국인',
    pop_seoul_columns[3] : '한국인남',
    pop_seoul_columns[4] : '한국인여',
    pop_seoul_columns[5] : '외국인',
    pop_seoul_columns[6] : '외국인남',
    pop_seoul_columns[7] : '외국인여'
},inplace=True)

pop_Seoul.head()
```

```python
pop_Seoul.drop([0],inplace=True) # 인덱스 0번의 데이터를 삭제하고 pop_Seoul 데이터 프레임에 적용
pop_Seoul.head()
```

```python
# nan 행 삭제
pop_Seoul.drop([26], inplace=True)
pop_Seoul.head()
```

```python
pop_Seoul['구별'].unique()
```

```python
len(pop_Seoul['구별'].unique())
```

```python
pop_Seoul['외국인남비율'] = (pop_Seoul['외국인남']/pop_Seoul['총인구'])*100
pop_Seoul['외국인녀비율'] = (pop_Seoul['외국인여']/pop_Seoul['총인구'])*100
pop_Seoul.head()
```

```python
pop_Seoul.sort_values(by='외국인남비율', ascending=False).head()
```

내림차순으로 정렬했을 때, 총 인구수 대비 외국인 남자 비율 top5는 영등포구, 금천구, 구로구, 용산구, 중구이다.

```python
pop_Seoul.sort_values(by='외국인녀비율', ascending=False).head()
```

내림차순으로 정렬했을 때, 총 인구수 대비 외국인 여자 비율 top5는 영등포구, 중구,금천구, 종로구, 구로구이다.
